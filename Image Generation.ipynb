```python
import random
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util
import numpy as np
from scipy import ndimage

def random_rotation(image_array: ndarray):
    # pick a random degree of rotation between 25% on the left and 25% on the right
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)

def random_noise(image_array: ndarray):
    # add random noise to the image
    return sk.util.random_noise(image_array)

def horizontal_flip(image_array: ndarray):
    # horizontal flip doesn't need skimage, it's easy as flipping the image array of pixels !
    return image_array[:, ::-1]
def change_contrast(original_image):
    v_min, v_max = np.percentile(original_image, (10.2, 79.8))
    better_contrast = sk.exposure.rescale_intensity(original_image, in_range=(v_min, v_max))
    return better_contrast
def gamma_correction(original_image):
    # gamma and gain parameters are between 0 and 1
    adjusted_gamma_image = sk.exposure.adjust_gamma(original_image, gamma=0.4, gain=0.9)
    return adjusted_gamma_image
def logarithmic_correction(original_image):
    log_correction_image = sk.exposure.adjust_log(original_image)
    return log_correction_image
def blur(original_image):
    blured_image = ndimage.uniform_filter(original_image, size=(5, 5, 1))
    return blured_image
```

```python
# dictionary of the transformations functions we defined earlier
available_transformations = {
    'gamma_correction':gamma_correction,
    'change_contrast':change_contrast,
    'noise': random_noise,
    'logarithmic_correction':logarithmic_correction,
    'rotate': random_rotation,
    'horizontal_flip': horizontal_flip,
    'blur':blur
}

# random num of transformations to apply

def generate(image_to_transform):
    num_transformations_to_apply = 1

    num_transformations = 0
    transformed_image = None
    while num_transformations <= num_transformations_to_apply:
        # choose a random transformation to apply for a single image
        key = random.choice(list(available_transformations))
        transformed_image = available_transformations[key](image_to_transform)
        new_file_path = '%s/augmented_image_%s.jpg' % (folder_path+'/augmented', num_generated_files)
        # write image to the disk
        image_to_transform = transformed_image
        num_transformations += 1
    sk.io.imsave(new_file_path, transformed_image)
```
